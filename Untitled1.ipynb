```python
import matplotlib
from matplotlib import pyplot as plt
x=[3,7,8]
y=[5,9,6]
x1=[2,4,6]
y1=[3,5,7]
plt.plot(x,y,'r',linewidth=2,labe1='line one')
plt.plot(x1,y1,'b',linewidth=6,label='line two')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Chart")
plt.legend()

plt.show()
```

```python
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[3,7,8]
y=[5,9,6]
x1=[2,4,6]
y1=[3,5,7]
plt.plot(x,y,'r',linewidth=2,label='line one')
plt.plot(x1,y1,'b',linewidth=6,label='line two')

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Chart")
plt.legend()
print(plt.style.available)
plt.show()
```

```python
import matplotlib
from matplotlib import pyplot as plt
import csv
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import style
style.use('dark_background')


fig=plt.figure()
ax = fig.add_subplot(111, projection='3d')
df=pd.read_csv(r"C:\Users\admin\PycharmProjects\tatvalabs\IRIS\iris.csv")
n=len(df)
for i in range(0,n):
    if str(df.iloc[i]['species'])=='setosa':
        ax.scatter(df.iloc[i]['sepal_length'],df.iloc[i]['sepal_width'],color='r')
    elif str(df.iloc[i]['species'])=='versicolor':
        ax.scatter(df.iloc[i]['sepal_length'], df.iloc[i]['sepal_width'], color='b')
    elif str(df.iloc[i]['species'])=='virginica':
        ax.scatter(df.iloc[i]['sepal_length'], df.iloc[i]['sepal_width'], color='g')

ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
plt.savefig('C:\\Users\\admin\\Desktop\\scatter.jpg')


plt.show()


```
